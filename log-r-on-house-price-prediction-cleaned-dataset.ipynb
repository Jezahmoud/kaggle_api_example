# Logistic Regression on House-Price-Prediction Cleaned Dataset

```python
#Importing all important libraries

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv('../input/housepriceprediction-cleaned-dataset/House-Price-Prediction-clean.csv')
```

# Data Description

```python
df.shape
```

```python
df.info()
```

```python
X = df.drop('SalePrice', axis=1)
```

```python
Y = df['SalePrice']
```

# Data Splitting on 80,20

```python
trainX, testX, trainY, testY = train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
#Logistic Regression model

modelLR = LogisticRegression()
```

```python
modelLR.fit(trainX, trainY)
```

```python
LRpredict = modelLR.predict(testX)
```

```python
mean_absolute_error(testY, LRpredict)
```

```python
mean_squared_error(testY, LRpredict)
```

```python
np.sqrt(mean_squared_error(testY, LRpredict))
```
